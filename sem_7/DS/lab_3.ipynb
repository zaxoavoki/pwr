```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-np.pi * 2, np.pi * 2, 200)
x2 = x1[:]

y1 = np.sin(x1) * np.exp(-(x1/5) ** 2)
y2 = np.cos(x2) * np.exp(-(x2/2) ** 2)

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x1, y1, 'b--', label='x1')
ax2.plot(x2, y2, 'r--', label='x2')

plt.xlabel('x label')
plt.ylabel('y label')
plt.ylim(-1.5, 1.5)
plt.xlim(-5, 5)
```

```python
# subplots

fig, axs = plt.subplots(2, 2)
fig.set_figheight(5)
fig.set_figwidth(15)

t = np.linspace(0, 1, 100)

axs[0, 0].plot(np.random.rand(40), np.random.rand(40), 'x', c='red')
axs[0, 1].plot(t ** 2, t ** 3 - t)
axs[1, 0].plot(np.random.rand(40))
axs[1, 1].plot(t * np.cos(10 * t), t * np.sin(10 * t))
```

```python
# 3d plotting
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-3, 3, 50)

X, Y = np.meshgrid(x, y)
Z = X ** 2 + X * Y ** 2 + Y ** 3

ax = Axes3D(plt.figure())
ax.plot_surface(X, Y, Z)
```

```python
# 3d plot
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)


X, Y = np.meshgrid(x, y)
z = np.cos(np.sqrt(X ** 2 + Y ** 2)) ** 2 * np.exp(-0.3 * np.sqrt(X ** 2 + Y ** 2))

ax = Axes3D(plt.figure())
ax.plot_surface(X, Y, z)
```

```python
t = np.linspace(0, 20, 500)
y = np.random.normal(1, 0.2, 500)

fig, axs = plt.subplots(3, 2)
fig.set_figwidth(15)
fig.set_figheight(8)

axs[0, 0].plot(t, sig.chirp(t, f0=4, f1=1, t1=10))
axs[0, 1].plot(t, y * sig.chirp(t, f0=4, f1=1, t1=10))
axs[1, 0].plot(t, sig.sawtooth(t))
axs[1, 1].plot(t, y * sig.sawtooth(t))
axs[2, 0].plot(t, sig.square(t))
axs[2, 1].plot(t, y * sig.square(t))
```

```python
t = np.linspace(0, 1, 250)
fs = 100
f, p = sig.periodogram(t, fs)
plt.plot(f, p)
```

```python
size = 100

t = np.linspace(-10, 10, size)
y = sig.sawtooth(t)
noice = np.random.normal(1, 0.5, size)

plt.plot(t, y * noice)
plt.plot(t, y)
plt.plot(t, np.sin(t), 'g--')
```

```python
size = 100

t = np.linspace(-10, 10, size)
y = np.sin(t)
noice = np.random.normal(1, 0.5, size)

Y = sig.wiener(y * noice, 15)

plt.plot(t, y * noice, 'c')
plt.plot(t, Y, 'r')
```

```python
t = np.linspace(0, 20, 500)
y = np.random.normal(1, 0.2, 500)

fig, axs = plt.subplots(3)
fig.set_figwidth(20)
fig.set_figheight(8)

x1 = y * sig.chirp(t, f0=4, f1=1, t1=10)
x2 = y * sig.sawtooth(t)
x3 = y * sig.square(t)

axs[0].plot(t, x1, t, sig.wiener(x1, 25))
axs[1].plot(t, x2, t, sig.wiener(x2, 25))
axs[2].plot(t, x3, t, sig.wiener(x3, 25))
```
